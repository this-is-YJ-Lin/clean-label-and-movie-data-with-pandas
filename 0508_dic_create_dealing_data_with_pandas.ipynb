```python
def create_dictione_one_key_one_value(text_path):
    dict_data_create = {}##UES
    dict_key = {}##TEM
    dict_key['whatever']="just_for_test"
    with open(text_path, encoding = 'utf8') as df:
        for kv in [d.strip().split('	') for d in df]:
            #print('kv[0](key)=',kv[0])
            if kv[0] not in dict_key['whatever']:
                dict_key['whatever'] = kv[0]
                #print('Now dic key:', dict_key)
            #print('kv[1](value)=',kv[1])
            dict_data_create.setdefault(kv[0], [])
            #print('dict_data_create_set default: ', dict_data_create)
            if len(dict_data_create[kv[0]])<1: 
                dict_data_create[kv[0]].append(kv[1])
                #print('if len(dict_data_create[kv[0]])<1 .... :',dict_data_create)
            #########add an if-else to deal with key: N/A , ONLY N/A NEED INSERTING EVERY VALUE
            else:
                if kv[0] == 'N/A':
                    dict_data_create[kv[0]].append(kv[1])
    return dict_data_create

dic_re = create_dictione_one_key_one_value('dic_pre_1-15422.txt')
```
