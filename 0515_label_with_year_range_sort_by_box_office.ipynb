```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv
import math
from math import log
from collections import Counter
```

```python
def create_dictione_one_key_one_value(text_path):
    dict_data_create = {}##Use
    dict_key = {}##Tem
    dict_key['whatever']="just_for_test"
    with open(text_path, encoding = 'utf8') as df:
        for kv in [d.strip().split('	') for d in df]:
            #print('kv[0](key)=',kv[0])
            if kv[0] not in dict_key['whatever']:
                dict_key['whatever'] = kv[0]
                #print('Now dic key:', dict_key)
            #print('kv[1](value)=',kv[1])
            dict_data_create.setdefault(kv[0], [])
            #print('dict_data_create_set default: ', dict_data_create)
            if len(dict_data_create[kv[0]])<1: 
                dict_data_create[kv[0]].append(kv[1])
                #print('if len(dict_data_create[kv[0]])<1 .... :',dict_data_create)
            #########add an if-else to deal with key: N/A , ONLY N/A NEED INSERTING EVERY VALUE
            else:
                if kv[0] == 'N/A':
                    dict_data_create[kv[0]].append(kv[1])
    return dict_data_create

dic_re = create_dictione_one_key_one_value('dic_pre_1-15422.txt')
```

```python

def label_by_year_sort_woth_box_office(year):
    
    group_year_movie = group_by_year.get_group(year)
    each_group_num = int(len(group_year_movie)/10)
    label_count = 9

    for i in range(0, len(group_by_year.get_group(year)),each_group_num):
        #group_year_movie.year_box_office_label.iloc[ i : i+each_group_num ] = label_count
        df_10038.year_box_office_label[group_by_year.groups[year][ i : i+each_group_num ]] = label_count
        label_count = label_count-1
        if df_10038.year_box_office_label[group_by_year.groups[year][i]] < 0:
            df_10038.year_box_office_label[group_by_year.groups[year][ i : i+each_group_num ]] = 0

            
            
for i in group_by_year.groups.keys():
    label_by_year_sort_woth_box_office(i)
```
